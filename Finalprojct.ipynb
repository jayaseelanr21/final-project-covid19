## 4. Filter the dataframe with only this columns
- ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.

```python
data = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
data
```

## 5. Data Cleaning
- Remove all duplicates observations
- Find missing values in all columns
- Remove all observations where continent column value is missing
- Tip : using subset parameter in dropna
- Fill all missing values with 0

```python
data.duplicated()
```

```python
data.isnull().sum()
```

```python
data1 = data.dropna()
```

```python
data1.isnull().sum()
```

## 6. Date time format :
- Convert date column in datetime format using pandas.to_datetime
- Create new column month after extracting month data from date column.

```python
data1.head()
```

```python
data1['date'] = pd.to_datetime(data1['date'],format = '%d/%m/%y',dayfirst = True)
data1.head()
```

```python
data1['month'] = data1['date'].dt.month
```

```python
data1.head()
```

## 7. Data Aggregation:
- Find max value in all columns using groupby function on 'continent' column
- Tip: use reset_index() after applying groupby
- Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)

```python
df_groupby = data1.groupby("continent").max().reset_index()
df_groupby
```

## 8. Feature Engineering :
- Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
data1['total_deaths_to_total_cases'] = data1['total_deaths']/data1["total_cases"]
data1.head()
```
